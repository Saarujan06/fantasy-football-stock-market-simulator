```python
import pandas as pd

# 1. Load the file (replace with the correct filename)
input_file = "cleaned_understat_epl_game_stats.csv"

df = pd.read_csv(input_file)

# 2. Convert the "season" column to numeric (it may be a string)
df["season"] = pd.to_numeric(df["season"], errors="coerce")

# 3. Filter out seasons before 2020
df_filtered = df[df["season"] >= 2020].copy()

# 4. Save cleaned file
output_file = "cleaned_game_stats_2020onwards.csv"
df_filtered.to_csv(output_file, index=False)

print("✅ Cleaned file saved as:", output_file)
print("Original rows:", len(df))
print("Rows from 2020 onwards:", len(df_filtered))
df_filtered.head()
```
