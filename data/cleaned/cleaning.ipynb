```python
import pandas as pd

# Replace with the path to your raw CSV file
input_file = "../raw/Prem 2020:2021.csv"


# Column names in the order they appear for EPL data from football-data.co.uk
column_names = [
    "Div", "Date", "Time", "HomeTeam", "AwayTeam",
    "FTHG", "FTAG", "FTR",
    "HTHG", "HTAG", "HTR",
    "Referee",
    "HS", "AS", "HST", "AST", "HF", "AF", "HC", "AC",
    "HY", "AY", "HR", "AR"
    # Betting odds columns come after this — we can ignore them
]

# Load the CSV with limited columns
df = pd.read_csv(input_file, header=None, names=column_names, usecols=column_names[:24], encoding='latin1')

# Drop unused columns
df_cleaned = df[[
    "Date", "HomeTeam", "AwayTeam", "FTHG", "FTAG", "FTR",
    "HS", "AS", "HST", "AST", "HF", "AF", "HC", "AC",
    "HY", "AY", "HR", "AR"
]].copy()

# Convert date format
df_cleaned["Date"] = pd.to_datetime(df_cleaned["Date"], dayfirst=True, errors='coerce')
df_cleaned = df_cleaned.dropna(subset=["Date"])

# Save cleaned version
output_file = "cleaned_match_data.csv"
df_cleaned.to_csv(output_file, index=False)

print(f"✅ Cleaned file saved as: {output_file}")
df_cleaned.head()
```
